## Underfitting VS Overfitting

```python
np.random.seed(0)

n_samples = 30
degrees = [2, 4, 15]

true_fun = lambda X: np.cos(1.5 * np.pi * X)

# Data Generation
X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1

plt.figure(figsize=(14, 5))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)
    plt.setp(ax, xticks=(), yticks=())

    polynomial_features = PolynomialFeatures(degree=degrees[i],
                                             include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    pipeline.fit(X[:, np.newaxis], y)

    # Evaluate the models using crossvalidation
    scores = model_selection.cross_val_score(pipeline,
        X[:, np.newaxis], y, scoring="neg_mean_squared_error", cv=10)

    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Model")
    plt.plot(X_test, true_fun(X_test), label="True function")
    plt.scatter(X, y, label="Samples")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})".format(
        degrees[i], -scores.mean(), scores.std()))
plt.show()
```

## Autoregression: build matrix of autoregression

```python
sc.linalg.hankel([1,2,3,4], [4,7,7,8,9])
```

```python
def get_autoregrmatrix(x,h,K):
    L = len(x)
    X = sc.linalg.hankel(x[:L-h-K+1], x[-K:])
    y = x[K+h-1:]
    return X,y
```

## Forecast for next hour (h=1) using 1 week history (K=168)

```python
K=168
h = 1 # delay

x_train = ts['EnergyCons'].ix['2005-01-01 00:00:00':'2006-01-01 00:00:00']
x_test = ts['EnergyCons'].ix['2006-01-01 00:00:00':'2006-01-30 00:00:00']

X_train, y_train = get_autoregrmatrix(x_train,h,K)
X_test, y_test = get_autoregrmatrix(x_test,h,K)
```

## Answer to questions 
  ** 1. What is time series? **
     
     Time series is a real-valued sequence which have time for argument.
       
  ** 2. What is delay and horizon of forecast?**
   
$\hspace{10mm}$ A period of time $D$ is called horizon of TS (for some pre-fixed model) if $D$ is the maximal time for which model can forecast value $\hat{y}_{T+D}$ of TS by having real values $y_1, \ldots, y_{T}$. Delay d is any point from start to $D$ (if time is counting by natural numbers, $d\in\{1,2,\ldots, D\}$). 
     
  ** 3. What is Least Squared Method for Linear Regression?**
  
$\hspace{10mm}$ By having the matrix $X = (x_{ij})_{N\times {p+1}} (N$ is the number of objects, $p$ is the number of features and $x_{i1} = 1$ for any $i\in \{1,2,\ldots, N\})$ and labels $Y=(y_1,\ldots, y_N)$, we can write the loss function 
$$
L(X,Y,W) = \sum_{n=1}^{N}(w_1x_{n1} + w_2x_{n2} + \ldots + w_{p+1}x_{n(p+1)}-y_{n})^2,
$$
where $W = (w_1, \ldots, w_{p+1})$ are called weigts. Least Squared Method is a method of finding the optimal weights by minimizing the loss function by weights.
          
  ** 4. What is the solution of linear regression via LS method (in terms of matrix $X, \alpha, Y$)? **
  
$\hspace{10mm}$ To find $\alpha = W$ it is sufficient to solve the equation $\frac{\partial}{\partial W}L = 0$. On the other hand we have 
$$
\frac{\partial}{\partial W}L =\frac{\partial}{\partial W}(XW-Y)(XW-Y)^T = 2X^T(XW-Y).
$$
So, the solution can be find from equation $X^TXW = X^TY$, and if $X^TX$ is invertible, we get $W = (X^TX)^{-1}X^TY$.
          
  ** 5. Write down main problems of LS method. **
  
$\hspace{10mm}$ The main problem is that the matrix $X^TX$ can be non-invertible and even if this matrix is invertible, it is too complicated to compute the inverse of $X^TX$.
          
  ** 6. What hankel transformation of a matrix? **
  
$\hspace{10mm}$ Let $\{b_n\}_{n=0}^{+\infty}$ a sequence. Hankel transformation of sequence $\{b_n\}_{n=0}^{+\infty}$ is the sequence $\{h_n\}_{n=0}^{+\infty}$, for which $h_n = \det((b_{i+j-2})_{1\leq i,j\leq n+1})$.
