2) 1. For a slower changing trend we will choose $\beta$ to be close to 0, because we can take into account all the previous data points with almost equal weight. 2. 3) **Proof the equivalency of correction error form and smoothing form for unobserved component of TS in  Thei-Wage model (period of seasonality equals $p$):**

  $$l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} )=l_{t-1} + b_{t-1} + \alpha e_t;$$
  $$ b_t = \beta (l_{t} - l_{t-1} ) + (1-\beta) b_{t-1} = b_{t-1} + \alpha\beta e_t;$$
  $$ s_t = \gamma (y_t-l_t) + (1-\gamma) s_{t-p} = s_{t-p} + \gamma(1-\alpha)e_t.$$
  
3) We get:

  $$ l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} ) = \alpha (e_t + l_{t-1} + b_{t-1} + s_{t-p} - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} ) = l_{t-1} + b_{t-1} + \alpha e_t. $$
  
Because we know that:
  $$ \hat{y}_{t|t-1} = l_{t-1} + b_{t-1} + s_{t-p};$$
  $$ e_t = y_t - \hat{y}_{t|t-1} = y_t - l_{t-1} + b_{t-1} + s_{t-p};$$
  thus $$y_t = e_t + l_{t-1} + b_{t-1} + s_{t-p}.$$  
  
Now we have that $l_t = l_{t-1} + b_{t-1} + \alpha e_t$, so, using the expression obtained for $l_t$ we can get:

$$ b_t = \beta (l_{t} - l_{t-1} ) + (1-\beta) b_{t-1} = \beta (l_{t-1} + b_{t-1} + \alpha e_t - l_{t-1} ) + (1-\beta) b_{t-1} = b_{t-1} + \alpha\beta e_t.$$

Also,
$$ s_t = \gamma (y_t-l_t) + (1-\gamma) s_{t-p} = \gamma ((e_t + l_{t-1} + b_{t-1} + s_{t-p})-(l_{t-1} + b_{t-1} + \alpha e_t)) + (1-\gamma) s_{t-p} = s_{t-p} + \gamma(1-\alpha)e_t.$$

Since we have obtained new expressions for $l_t$ and $y_t$.


4) **Find the correction error form for unobserved component of TS in ES with additive damped trend:**

$$\hat{y}_{t+d|t} = l_t + \left(\phi + \phi^2 + \dots + \phi^{d}\right) b_t.$$

**Unobserved components are (in smoothing form): **

$$ l_t       = \alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right),$$
$$ b_t         = \beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1}.$$


4) $$ l_t = \alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right) = \alpha (e_t + l_{t-1} + \phi b_{t-1}) + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right) = \alpha e_t + l_{t-1} +\phi b_{t-1}. $$

Because similar to $\hat{y}_{t+d|t}$, we can get that:

$$\hat{y}_{t|t-1} = l_{t-1} + \phi b_{t-1};$$
$$ e_t = y_t - \hat{y}_{t|t-1} = y_t - l_{t-1} + \phi b_{t-1};$$
thus
$$y_t = e_t + l_{t-1} + \phi b_{t-1} .$$

Now that we have a new expression for $l_t$, we can get a new expression for $b_t$ by a simple substitution:

$$ b_t = \beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1} = \beta \left(\alpha e_t + l_{t-1} +\phi b_{t-1} - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1} = \beta \alpha e_t + \phi b_{t-1}.$$
